```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function_1(x, option: int):
    if option == 1:
        return (x**2 - 1) / (x - 1)
    elif option == 2:
        return (x**2 - 25) / (x + 5)
    elif option == 3:
        return (x**2 - 2*x - 8) / (x - 4)
    elif option == 4:
        return 25 / x

def plot_graph(x_value, y_value, xlim: tuple, ylim: tuple, xaxvline: int, yaxhline: int):
    fig, ax = plt.subplots()
    plt.xlim(xlim[0], xlim[1])
    plt.ylim(ylim[0], ylim[1])
    plt.axvline(x=xaxvline, color='purple', linestyle='--')
    plt.axhline(y=yaxhline, color='purple', linestyle='--')
    return ax.plot(x_value, y_value)
```

```python
x = np.linspace(-10, 10, 10000)
```

$$ \lim_{x \to 0} \frac{x^2-1}{x-1} $$

```python
y = function_1(x, 1)
print(function_1(0, option=1))
_ = plot_graph(x, y, (-10, 10), (-2, 4), 0, 1)
```

$$ \lim_{x \to -5} \frac{x^2-25}{x+5} $$

```python
y = function_1(x, 2)
print(function_1(-4.999999, option=2))
_ = plot_graph(x, y, (-10, 20), (-20, 20), -5, -10)
```

$$ \lim_{x \to 4} \frac{x^2 -2x -8}{x-4} $$

```python
y = function_1(x, 3)
print(function_1(3.99999999, option=3))
print(function_1(4.001, option=3))
_ = plot_graph(x, y, (-6, 6), (-10, 10), 4, 6)
```
