```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sbn

%matplotlib inline
```

```python
data = pd.read_csv("haberman.csv")
data.shape
```

```python
data.columns
```

```python
data.head(10)
```

```python
data['status'].unique()
```

```python
data_survived = data[data.status == 1]
data_dead = data[data.status == 2]

print(data_survived.shape)
print(data_dead.shape)
```

```python
data['status'].value_counts()
```

```python
data.plot(kind = 'scatter', x = 'nodes', y = 'status')
plt.show()
```

```python
data.plot(kind = 'scatter', x = 'age', y = 'status')
plt.show()
```

```python
sbn.set_style("whitegrid")
sbn.FacetGrid(data, hue='status', height=5) \
    .map(plt.scatter, "age", "nodes") \
    .add_legend()

plt.show()
plt.close()
```

```python
sbn.set_style("whitegrid")
sbn.pairplot(data, hue = "status", height = 4, palette='Dark2')
plt.show()
plt.close()
```

```python
sbn.FacetGrid(data, hue='status', height=6) \
    .map(sbn.histplot, "nodes") \
    .add_legend()

plt.show()
plt.close()
```

```python
sbn.FacetGrid(data, hue='status', height=6) \
    .map(sbn.histplot, "age") \
    .add_legend()

plt.show()
plt.close()
```

```python
sbn.FacetGrid(data, hue='status', height=6) \
    .map(sbn.distplot, "nodes") \
    .add_legend()

plt.show()
plt.close()
```

```python
sbn.FacetGrid(data, hue='status', height=6) \
    .map(sbn.distplot, "age") \
    .add_legend()

plt.show()
plt.close()
```
