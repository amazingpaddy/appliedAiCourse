```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
iris = pd.read_csv("iris.csv")
```

```python
print(iris.shape)
```

```python
iris.head(5)
```

```python
print(iris.columns)
```

```python
iris["species"].value_counts()
```

```python
iris.plot(kind = 'scatter', x = 'sepal_length', y = 'sepal_width')
plt.show()
```

```python
sb.set_style("whitegrid")
sb.FacetGrid(iris, hue='species', height=5)\
    .map(plt.scatter, "sepal_length", "sepal_width")\
    .add_legend()

plt.show()
```

```python
plt.close()
sb.set_style("whitegrid")
sb.pairplot(iris, hue="species", height=4)
plt.show()
```

```python
plt.close()
sb.FacetGrid(iris, hue = "species", height = 6)\
    .map(sb.histplot, "petal_length")\
    .add_legend();
plt.show();
```

```python
sb.FacetGrid(iris, hue="species", height=5)\
    .map(sb.distplot, "petal_length")\
    .add_legend()
plt.show()
```

```python
sb.FacetGrid(iris, hue="species", height=5)\
    .map(sb.distplot, "petal_width")\
    .add_legend()
plt.show()
```

```python
sb.FacetGrid(iris, hue="species", height=5)\
    .map(sb.distplot, "sepal_length")\
    .add_legend()
plt.show()
```

```python
print(iris["species"].unique())
```

```python
iris_sen = iris[iris.species == "Iris-setosa"]
iris_vir = iris[iris.species == "Iris-virginica"]
iris_ver = iris[iris.species == "Iris-versicolor"]
```

```python
print(np.mean(iris_sen["petal_length"]))
print(np.mean(iris_vir["petal_length"]))
print(np.mean(iris_ver["petal_length"]))
```

```python
print(np.std(iris_sen["petal_length"]))
print(np.std(iris_vir["petal_length"]))
print(np.std(iris_ver["petal_length"]))
```

```python
print(np.median(iris_sen["petal_length"]))
print(np.median(iris_vir["petal_length"]))
print(np.median(iris_ver["petal_length"]))

print("Appending outlier")

print(np.median(np.append(iris_vir["petal_length"], 50))) #Appending outlier, but median wont change.
```
